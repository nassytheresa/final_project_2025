```python
from pycoingecko import CoinGeckoAPI
import pandas as pd

# Initialize the CoinGeckoAPI client
cg = CoinGeckoAPI()

# Prepare to collect market data
all_coins_market = []

# Loop through pages to fetch data
for page in range(1, 6):  # Adjust the range as needed
    parameters = {
        'vs_currency': 'usd',
        'order': 'market_cap_desc',
        'per_page': 250,
        'page': page,
        'sparkline': False,
        'locale': 'en'
    }
    market_data = cg.get_coins_markets(**parameters)
    all_coins_market.extend(market_data)

# Convert the collected data into a DataFrame
df_all_coins = pd.DataFrame(all_coins_market)

# Display the first few rows of the DataFrame
print(df_all_coins.head())
```
